```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import matplotlib as mpl
import textwrap
```

```python
rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = 'white'

colors = {
    'orange':'#f4731f',
    'turquoise': '#41B6E6',
    'purple': '#512D6D',
    'blue': '#006298',
    'l_blue': '#99D6EA',
    'd_blue': '#01426A',
    'green': '#78BE20'
}
```

```python
data_tot = pd.read_csv('../data/q1_df.csv', usecols = ['label', 'know_someone', 'experienced'])
```

```python
#data_tot = data_tot[['label', 'know_someone', 'experienced']].copy()
data_tot['know_exp'] = data_tot['know_someone'] + data_tot['experienced']
```

```python
data_tot
```

```python
data = pd.read_csv('../data/q1_child_lt13_df.csv')
data.drop(index = [0, 1], inplace = True)
data.drop(columns = ['freq_1', 'freq_2', 'freq_3', 'freq_4', 'freq_5'], inplace = True)
data = data[['prop_1', 'prop_2', 'prop_3', 'prop_4', 'prop_5']].T.copy()
data.columns = ['know_someone', 'experienced']
data['know_exp'] = data['know_someone'] + data['experienced']
data.reset_index(drop = True, inplace = True)
data = data.join(data_tot['label'])
```

```python
data
```

```python
data_tot.sort_values('know_exp', inplace = True)
data_tot.reset_index(drop = True, inplace = True)
data.sort_values('know_exp', inplace = True)
data.reset_index(drop = True, inplace = True)
data[['know_someone', 'experienced', 'know_exp']] = round(data[['know_someone', 'experienced', 'know_exp']]).astype('int').copy()
```

```python
data_tot
```

```python
data
```

```python
fig, ax = plt.subplots(ncols = 2, figsize=(7.5,5), sharey = True, sharex = True)

know = "I know someone who has experienced this recently"
exp = "I have personally experienced this recently"

fontsize_yticks = 12
fontsize_title = 20
fontsize_subtitle = 12
fontsize_datalabs = 10
fontsize_caption = 8

fig.suptitle("Majorities with young children know someone or\nhave experienced childcare challenges".upper(), 
             ha='center', va='top', fontsize=fontsize_title, fontproperties = 'DIN Condensed')

ax[0].set_xticks([])
ax[0].set_xlim((0, 78))
ax[0].set_yticks(data_tot.index)
ax[0].set_yticklabels(labels=data_tot['label'].apply(lambda x: textwrap.fill(x, 50)), 
                      fontsize=fontsize_yticks, color = text_color)

ax[0].set_title("Total".upper(),
                fontsize = fontsize_subtitle)
ax[0].barh(data_tot.index, data_tot['experienced'], label = exp, color = colors['turquoise'])
ax[0].barh(data_tot.index, data_tot['know_someone'], left = data_tot['experienced'], label = know, 
           color = colors['turquoise'], alpha=0.5)
for x,y in zip(data_tot['experienced'], data_tot.index):
    label = f"{x}%"
    ax[0].annotate(label,
                   ((x/2)+1,y),
                   ha='center', va='center',
                   fontsize=fontsize_datalabs, #color = 'white', 
                   weight = 'bold')
for x,y,z in zip(data_tot['know_someone'], data_tot.index, data_tot['experienced']):
    label = f"{x}%"
    ax[0].annotate(label,
                   (z+(x/2),y),
                   ha='center', va='center',
                   fontsize=fontsize_datalabs, #color = 'white', 
                   weight = 'bold')
for x,y in zip(data_tot['know_exp'], data_tot.index):
    label = f"{x}%"
    ax[0].annotate(label,
                   (x+8,y),
                   ha='center', va='center',
                   fontsize=fontsize_datalabs, weight = 'bold')

ax[1].set_title("Parent w/\nyoung child".upper(),
                fontsize = fontsize_subtitle)
ax[1].barh(data.index, data['experienced'], label = exp, color = colors['turquoise'])
ax[1].barh(data.index, data['know_someone'], left = data['experienced'], label = know, 
           color = colors['turquoise'], alpha=0.5)
for x,y in zip(data['experienced'], data.index):
    label = f"{x}%"
    ax[1].annotate(label,
                   (x/2,y),
                   ha='center', va='center',
                   fontsize=fontsize_datalabs, #color = 'white', 
                   weight = 'bold')
for x,y,z in zip(data['know_someone'], data.index, data['experienced']):
    label = f"{x}%"
    ax[1].annotate(label,
                   (z+(x/2),y),
                   ha='center', va='center',
                   fontsize=fontsize_datalabs, #color = 'white', 
                   weight = 'bold')
for x,y in zip(data['know_exp'], data.index):
    label = f"{x}%"
    ax[1].annotate(label,
                   (x+8,y),
                   ha='center', va='center',
                   fontsize=fontsize_datalabs, weight = 'bold')


sns.despine(bottom=True)
fig.tight_layout(rect = [0, 0.2, 1, 1])

handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, ncol = 2, 
           prop={'size':10}, frameon=False, handlelength = 1, 
           loc = 'upper center', bbox_to_anchor = (0.5,0.2))

# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q1. Sample: 2,052 US adults. Dates: August 9-11, 2021.
Subsample of n=522 (unweighted) respondents with a child age 12 or younger.
See topline questionnaire in appendix for question wording.'''
fig.text(x = 0, y = 0, 
         ha = 'left', va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':fontsize_caption})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.9,0,0.1,0.1], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

# save image file
#plt.show();
fig.savefig('../images/fig04.png', transparent = True, dpi=300, bbox_inches='tight');
```
