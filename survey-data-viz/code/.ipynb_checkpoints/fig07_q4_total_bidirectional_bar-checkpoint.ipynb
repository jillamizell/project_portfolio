```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib import rcParams
import ast
```

```python
rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.labelcolor'] = text_color
mpl.rcParams['ytick.major.width'] = 0

colors = {
    'orange':'#f4731f',
    'turquoise': '#41B6E6',
    'purple': '#512D6D',
    'blue': '#006298',
    'l_blue': '#99D6EA',
    'd_blue': '#01426A',
    'green': '#78BE20'
}
```

```python
data = pd.read_csv('../data/q4_df.csv')
```

```python
data['resp_1'] = data['resp_1']*-1
```

```python
data
```

```python
data_tot = data[data['group'] == 'total'].copy()
data_tot.drop(columns = 'group', inplace = True)
data_tot.sort_values('resp_0', inplace = True)
data_tot.set_index('quest', inplace = True)
```

```python
data_tot[['resp_0', 'resp_1']].abs().max().max()
```

```python
max_lim = data_tot[['resp_0', 'resp_1']].abs().max().max() + 20

fontsize_yticks = 12
fontsize_title = 20
fontsize_datalabs = 12
fontsize_caption = 8
nudge_datalab = 8

# instantiate subplot objects
fig, ax = plt.subplots(ncols = 2, figsize = (7.5,3.5), sharey = True)

# plot title
fig.suptitle('Performance of U.S. Childcare System'.upper(),
             fontsize = fontsize_title,
             ha='center', va = 'top',
             fontproperties = 'DIN Condensed',
             weight = 'bold')


# remove lines around plot
sns.despine(bottom=True, left=True) 

# dimmensions and customizations for the negative (left) bar plot - TOTAL
ax[0].axis('off')
ax[0].set_xticks([])
ax[0].set_xlim([max_lim * -1,0])  
ax[0].barh(data_tot.index, data_tot['resp_1'], color = list(colors.values())[:4][::-1], alpha = 0.5)

# create data value labels
for x,y in zip(data_tot['resp_1'], data_tot.index):
    label = f"{(x*-1)}%"
    ax[0].annotate(label, 
                   (x-nudge_datalab,y), 
                   ha='center', 
                   va='center',
                   fontsize=fontsize_datalabs,
                   weight = 'bold')
    y_lab = ast.literal_eval(data_tot['quest_resp'].loc[y])[1]
    ax[0].annotate(y_lab,
                   (-96, y),
                   ha = 'right', va = 'center',
                   fontsize = fontsize_yticks)


# dimmensions and customizations for the positive (right) bar plot - TOTAL
ax[1].axis('off')
ax[1].set_yticks(range(0, len(data_tot.index)))
ax[1].yaxis.tick_right()
ax[1].barh(data_tot.index, data_tot['resp_0'], color = list(colors.values())[:4][::-1])
ax[1].set_xlim([0,max_lim]) 
ax[1].set_xticks([]) 
# create data value labels
for x,y in zip(data_tot['resp_0'], data_tot.index):
    label = f"{(x)}%"
    ax[1].annotate(label, 
                 (x+nudge_datalab,y), 
                 ha='center', 
                 va='center',
                 fontsize=fontsize_datalabs,
                 weight = 'bold')
    y_lab = ast.literal_eval(data_tot['quest_resp'].loc[y])[0]
    ax[1].annotate(y_lab,
                   (83, y),
                   ha = 'left', va = 'center',
                   fontsize = fontsize_yticks)

# tight layout
fig.tight_layout(rect = [0, 0.1, 1, 0.9], pad=0)
    
# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q4. Sample: 2,052 US adults. Dates: August 9-11, 2021. 
See topline questionnaire in appendix for question wording.'''
fig.text(x = 0, y = 0, 
         ha = 'left', va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':fontsize_caption})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.9,0,0.1,0.1], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

# save image file
#plt.show();
fig.savefig('../images/fig07.png', facecolor = "white", dpi=300, bbox_inches='tight');
```
