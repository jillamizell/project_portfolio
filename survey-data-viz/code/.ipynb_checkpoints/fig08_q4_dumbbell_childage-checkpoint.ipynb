```python
l_blue = "#41B6E6"
d_blue = "#006298"
purple = '#512D6D'
orange = '#f4731f'
grey = '#323232'
color_list = [orange, l_blue, purple, d_blue]

var_list = list(data_yc.index)[::-1]

my_dict = {
    0: {'label': 'Quality', 'sub': 'none', 'quest': var_list[0]}, 
    1: {'label': 'Parent of child 12 or younger', 'sub': 'yc', 'quest': var_list[0], 'color': color_list[0], 
        'font_color_0': grey, 'font_color_1': grey}, 
    2: {'label': 'Parent of child 13 or older', 'sub': 'oc', 'quest': var_list[0], 'color': color_list[0], 
        'font_color_0': grey, 'font_color_1': grey}, 
    3: {'label': 'No children', 'sub': 'nc', 'quest': var_list[0], 'color': color_list[0], 
        'font_color_0': grey, 'font_color_1': grey},
    4: {'label': 'Distance', 'sub': 'none', 'quest': var_list[1]}, 
    5: {'label': 'Parent of child 12 or younger', 'sub': 'yc', 'quest': var_list[1], 'color': color_list[1], 
        'font_color_0': grey, 'font_color_1': grey},
    6: {'label': 'Parent of child 13 or older', 'sub': 'oc', 'quest': var_list[1], 'color': color_list[1], 
        'font_color_0': grey, 'font_color_1': grey},
    7: {'label': 'No children', 'sub': 'nc', 'quest': var_list[1], 'color': color_list[1], 
        'font_color_0': grey, 'font_color_1': grey},
    8: {'label': 'Access', 'sub': 'none', 'quest': var_list[2]}, 
    9: {'label': 'Parent of child 12 or younger', 'sub': 'yc', 'quest': var_list[2], 'color': color_list[2], 
        'font_color_0': grey, 'font_color_1': grey},
    10: {'label': 'Parent of child 13 or older', 'sub': 'oc', 'quest': var_list[2], 'color': color_list[2], 
         'font_color_0': 'white', 'font_color_1': grey},
    11: {'label': 'No children', 'sub': 'nc', 'quest': var_list[2], 'color': color_list[2], 
         'font_color_0': 'white', 'font_color_1': grey},
    12: {'label': 'Cost', 'sub': 'none', 'quest': var_list[3]}, 
    13: {'label': 'Parent of child 12 or younger', 'sub': 'yc', 'quest': var_list[3], 'color': color_list[3], 
         'font_color_0': 'white', 'font_color_1': grey},
    14: {'label': 'Parent of child 13 or older', 'sub': 'oc', 'quest': var_list[3], 'color': color_list[3], 
         'font_color_0': 'white', 'font_color_1': grey},
    15: {'label': 'No children', 'sub': 'nc', 'quest': var_list[3], 'color': color_list[3], 
         'font_color_0': 'white', 'font_color_1': grey},
}

# font and point sizes, locations
fontsize_yticks = 12
fontsize_title = 20
fontsize_datalabs = 12
fontsize_caption = 8

point_size = 1000
hline_adjust = 3

# universal axis settings
xmin = 10
xmax = 90
ymin = -0.5
ymax = 0.5  


# create figure and axes
fig, ax = plt.subplots(nrows = len(my_dict), sharex = True, figsize = (7.5,len(my_dict)-6))

title = "Performance of U.S. Childcare System, by parental status and age of child"
fig.suptitle(title.upper(), 
             ha='center', va = 'top',
             fontsize=fontsize_title, fontproperties = 'DIN Condensed')

# x-axis settings
ax[0].set_xticks([])
ax[0].set_xlim(xmin = xmin, xmax = xmax)

# loop
for i in range(0,len(my_dict)): 
    ax[i].set_yticks([0]) 
    ax[i].set_ylim(ymin = ymin, ymax = ymax)
    if my_dict[i]['sub'] == 'none':
        ax[i].set_yticklabels(labels = [my_dict[i]['label'].upper()], fontsize = fontsize_yticks+4, weight = 'bold')
        if data_yc.loc[my_dict[i]['quest']]['resp_0']>data_yc.loc[my_dict[i]['quest']]['resp_1']:
            set_ha_0 = 'left'
            set_ha_1 = 'right'
        else:
            set_ha_0 = 'right'
            set_ha_1 = 'left'
        ax[i].annotate(ast.literal_eval(data_yc.loc[my_dict[i]['quest']]['quest_resp'])[0],
               (data_yc.loc[my_dict[i]['quest']]['resp_0'], -0.1),
               ha = set_ha_0, 
               va = 'top',
               fontsize = fontsize_yticks)
        ax[i].annotate(ast.literal_eval(data_yc.loc[my_dict[i]['quest']]['quest_resp'])[1],
               (data_yc.loc[my_dict[i]['quest']]['resp_1'], -0.1),
               ha = set_ha_1, 
               va = 'top',
               fontsize = fontsize_yticks)
    else:
        ax[i].set_yticklabels(labels = [my_dict[i]['label']], fontsize = fontsize_yticks)
        if my_dict[i]['sub'] == 'yc':
            sub = data_yc
        elif my_dict[i]['sub'] == 'oc':
            sub = data_oc
        elif my_dict[i]['sub'] == 'nc':
            sub = data_nc
        ax[i].scatter(x = sub.loc[my_dict[i]['quest']]['resp_0'], 
                      y = 0,
                      color=my_dict[i]['color'], 
                      alpha=1, 
                      s=point_size)
        ax[i].scatter(x = sub.loc[my_dict[i]['quest']]['resp_1'], 
                      y = 0,
                      color=my_dict[i]['color'], 
                      alpha=0.5, 
                      s=point_size)
        if np.abs(sub.loc[my_dict[i]['quest']]['resp_0'] - sub.loc[my_dict[i]['quest']]['resp_1']) > 5:
            ax[i].hlines(y = 0,
                         xmin=sub.loc[my_dict[i]['quest']][['resp_0', 'resp_1']].min()+hline_adjust, 
                         xmax=sub.loc[my_dict[i]['quest']][['resp_0', 'resp_1']].max()-hline_adjust, 
                         color='grey', 
                         alpha=0.4,
                         lw = 5)
            ax[i].annotate(f"{sub.loc[my_dict[i]['quest']]['resp_1']}%",
                           (sub.loc[my_dict[i]['quest']]['resp_1'], 0),
                           ha = 'center', 
                           va = 'center', 
                           fontsize = fontsize_datalabs,
                           color = my_dict[i]['font_color_1'],
                           weight = 'bold')
            ax[i].annotate(f"{sub.loc[my_dict[i]['quest']]['resp_0']}%",
                           (sub.loc[my_dict[i]['quest']]['resp_0'], 0),
                           ha = 'center', 
                           va = 'center', 
                           fontsize = fontsize_datalabs,
                           color = my_dict[i]['font_color_0'],
                           weight = 'bold')

ax[9].annotate(f"{data_yc.loc[my_dict[9]['quest']]['resp_1']}%",
               (data_yc.loc[my_dict[9]['quest']]['resp_1']-6, 0),
               ha = 'center', 
               va = 'center', 
               fontsize = fontsize_datalabs,
               color = grey,
               weight = 'bold')
ax[9].annotate(f"{data_yc.loc[my_dict[9]['quest']]['resp_0']}%",
               (data_yc.loc[my_dict[9]['quest']]['resp_0']+6, 0),
               ha = 'center', 
               va = 'center', 
               fontsize = fontsize_datalabs,
               color = grey,
               weight = 'bold')

# tight layout
fig.tight_layout(rect = [0, 0.1, 1, 0.95], pad=0)

# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q4. Sample: 2,052 US adults. Dates: August 9-11, 2021.
See topline questionnaire in appendix for question wording.'''
fig.text(x = 0, y = 0.05, 
         ha = 'left', va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':fontsize_caption})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0,0,0.1,0.1], anchor='SW', zorder=1)
newax.imshow(logo)
newax.axis('off')

sns.despine(bottom = True, left = True)
#plt.show();
fig.savefig('../images/fig08.png', transparent = True, dpi=300, bbox_inches='tight');
```
