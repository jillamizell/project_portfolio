```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import matplotlib as mpl 
import textwrap
```

```python
rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = 'white'

colors = {
    'orange':'#f4731f',
    'turquoise': '#41B6E6',
    'purple': '#512D6D',
    'blue': '#006298',
    'l_blue': '#99D6EA',
    'd_blue': '#01426A',
    'green': '#78BE20'
}
```

```python
with open('../data/q5.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
# get sorted index based on total values
total = data['total'].copy()
total.reset_index(inplace = True)
labels = dict(zip(total['q5_grpd'], total['Labels']))
total.sort_values('Prop', ascending = True, inplace = True)
total.reset_index(drop = True, inplace = True)
total['Prop'] = round(total['Prop']).astype('int')
total
```

```python
# combine relevant information into one dataframe
var_list = list(data.keys())[1:]
tot_prop = data['total'][['Prop']].astype('int')
grouping_var = ['total']
group = ['total']
gov = list(tot_prop.loc[1])
emp = list(tot_prop.loc[2])
sup_sys = list(tot_prop.loc[3])
par_res = list(tot_prop.loc[4])
dk = list(tot_prop.loc[5])

for var in var_list:
    test = data[var].astype('int')
    grouping_var += [var]*len(test)
    group += list(test.index)
    gov += list(test[1])
    emp += list(test[2])
    sup_sys += list(test[3])
    par_res += list(test[4])
    dk += list(test[5])

all_df = pd.DataFrame({
    'grouping_var': grouping_var,
    'group': group,
    'gov': gov,
    'emp': emp,
    'sup_sys': sup_sys,
    'par_res': par_res,
    'dk': dk
})
```

```python
# filter dataframe to just keep relevant information
var_list = ['child_net_lt13', 'child_net_13up', 'child_none', 'age_cat', 'residential_area', 
            'employment_status', 'race_white_nh', 'race_hisp', 'race_black_nh']
# filter out rows for variables i don't want
sub_df = all_df[all_df['grouping_var'].isin(var_list)].copy()
sub_df.drop(index = sub_df[sub_df['group'] == 'No'].index, inplace = True)
sub_df.loc[sub_df['group'] == 'Yes', 'group'] = sub_df.loc[sub_df['group'] == 'Yes', 'grouping_var']
```

```python
# making the plot rows the index 
sub_df['ord'] = [1, 2, 3, 5, 6, 7, 8, 9, 10, 15, 16, 17, 12, 13, 21, 19, 20]
sub_df.sort_values('ord', inplace = True)
sub_df.set_index('ord', inplace = True)

# correct grouping)var
sub_df.loc[1, 'grouping_var'] = 'parental_status'
sub_df.loc[2, 'grouping_var'] = 'parental_status'
sub_df.loc[3, 'grouping_var'] = 'parental_status'
sub_df.loc[19, 'grouping_var'] = 'race_ethnicity'
sub_df.loc[20, 'grouping_var'] = 'race_ethnicity'
sub_df.loc[21, 'grouping_var'] = 'race_ethnicity'
# correct group labels
sub_df.loc[1, 'group'] = 'Parent of child age 12 or under'
sub_df.loc[2, 'group'] = 'Parent of child age 13 or older'
sub_df.loc[3, 'group'] = 'No children'
sub_df.loc[12, 'group'] = 'Employed'
sub_df.loc[13, 'group'] = 'Not employed'
sub_df.loc[19, 'group'] = 'Hispanic'
sub_df.loc[20, 'group'] = 'Black'
sub_df.loc[21, 'group'] = 'White'
```

```python
subtitles = {
    0: "Age of children",
    4: "Age",
    11: "Employment status",
    14: "Residential area",
    18: "Race/Ethnicity"
}
```

```python
nrows = sub_df['grouping_var'].nunique() + len(sub_df)
row_range = range(0, nrows)
ytick_labs = []
for r in row_range:
    if r in list(sub_df.index):
        ytick_labs += [sub_df.loc[r, 'group']]
    else:
        ytick_labs += [subtitles[r]]
```

```python
col_dict = {
    1: {'colname': 'gov', 'color':colors['l_blue']},
    2: {'colname': 'emp', 'color':colors['turquoise']},
    3: {'colname': 'sup_sys', 'color':colors['blue']},
    4: {'colname': 'par_res', 'color':colors['purple']},
    5: {'colname': 'dk', 'color':colors['orange']}
}
```

```python
c=1
textwrap.fill(labels[col_ord[c]], 15)
```

```python
col_ord = {0:1, 1:3, 2:4, 3:2, 4:5}
```

```python
subtitles.keys()
```

```python
# create plot objects

nrows = sub_df['grouping_var'].nunique() + len(sub_df)
ncols = len(labels)

fig, ax = plt.subplots(nrows = nrows, ncols=ncols, 
                       figsize=(15,15), sharex=True)#, sharey=True)

# font and point sizes, locations
caption_size = 20
ytick_lab_size = 20
data_lab_size = 18
sub_title_size = 20

# universal axis settings
xmin = 0
xmax = 40
ymin = -0.5
ymax = 0.5  

title = "Institution with a primary role in supporting parents with childcare challenges"
fig.suptitle(title.upper(), 
#             x = .32, y = 0.95,
             ha='center', va='center', fontsize=34, fontproperties = 'DIN Condensed')

# axis settings
ax[0,0].set_xticks([])
ax[0,0].set_xlim(xmin = xmin, xmax = xmax)

# loop
col_range = range(0, ncols)
row_range = range(0, nrows)

# ax[0,0].set_yticklabels(labels = [ytick_labs[0]])

for c in col_range:
    temp_dict = col_dict[col_ord[c]]
    temp_title = textwrap.fill(labels[col_ord[c]], 15)
    ax[0,c].set_title(temp_title, ha='center', va='bottom', fontsize=sub_title_size)
    for r in row_range:
        ax[r,c].set_ylim(ymin = ymin, ymax = ymax)
        ax[r,c].set_yticks([0])
        #ax[r,c].yaxis.set_visible(False) 
        if c == 0:
            if r in list(subtitles.keys()):
                ax[r,c].set_yticklabels(labels = [ytick_labs[r].upper()], fontsize = ytick_lab_size, 
                                        color = text_color, weight = 'bold')
            else:
                ax[r,c].set_yticklabels(labels = [ytick_labs[r]], fontsize = ytick_lab_size, color = text_color)
        else:
            ax[r,c].set_yticklabels([])
        if r in list(sub_df.index):
            x = sub_df[temp_dict['colname']].loc[r]
            if x > 9:
                x_nudge = 5
            else:
                x_nudge = 3
            ax[r,c].barh(y=0, width=x, color = temp_dict['color'])
            ax[r,c].annotate(f"{x}%", (x-x_nudge,0),
                             ha='center', va='center', fontsize=data_lab_size, color = 'white', weight = 'bold')  
ax[0,4].set_title("Not sure/\nsomething else", ha='center', va='bottom', fontsize=sub_title_size)

# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q5. Sample: 2,052 US adults. Dates: August 9-11, 2021. 
See topline questionnaire in appendix for question wording.'''
fig.text(x = 0, y = 0, 
         ha = 'left',
         va = 'bottom',
         ma = 'left',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':caption_size})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.9,0,0.1,0.1], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

fig.tight_layout(h_pad = 0, rect = (0, 0.05, 1, 1))
sns.despine(left = True, bottom = True)
```
