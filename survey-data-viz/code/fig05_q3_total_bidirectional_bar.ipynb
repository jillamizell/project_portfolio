```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib import rcParams
import textwrap
```

```python
rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.labelcolor'] = text_color
mpl.rcParams['ytick.major.width'] = 0

colors = {
    'orange':'#f4731f',
    'turquoise': '#41B6E6',
    'purple': '#512D6D',
    'blue': '#006298',
    'l_blue': '#99D6EA',
    'd_blue': '#01426A',
    'green': '#78BE20'
}
```

```python
data = pd.read_csv('../data/q3_tot_yc_df.csv')
data['poor'] = -1 * data['poor']
```

```python
wording = {
    'q3_1_grpd': "Parents who work non-standard\nhours (e.g., third shift)",
    'q3_2_grpd': "Parents who work standard\nhours (e.g., 9 to 5)",
    'q3_3_grpd': "Children",
    'q3_4_grpd': "Childcare workers"
}
```

```python
data_tot = data[data['group'] == 'total'].copy()
data_tot.drop(columns = 'group', inplace = True)
data_tot.set_index('quest', inplace = True)
```

```python
data_yc = data[data['group'] == 'young_child'].copy()
data_yc.sort_values('good', inplace = True)
data_yc.drop(columns = 'group', inplace = True)
data_yc.set_index('quest', inplace = True)
data_tot = data_tot.reindex(list(data_yc.index))
```

```python
list(data_yc.index)
```

```python
[(wording[k]) for k in list(data_yc.index)]
```

```python
l_blue = "#41B6E6"
d_blue = "#006298"

fontsize_yticks = 12
fontsize_title = 20
fontsize_subtitle = 12
fontsize_datalabs = 12
fontsize_caption = 8
nudge_datalab = 14
max_xaxis = 90

subtitle_y = 0.8

# instantiate subplot objects
fig, ax = plt.subplots(ncols = 4, sharey=True, figsize = (7.5,4))

# plot title
fig.suptitle('Performance of U.S. Childcare System'.upper(),
             fontsize = fontsize_title,
             ha='center', va = 'top',
             fontproperties = 'DIN Condensed',
             weight = 'bold')

# add "Total" and "Parent" subtitles
fig.text(x = .49, y = subtitle_y, 
         ha = 'center', va = 'bottom',
         s = "Total".upper(), 
         fontdict={'fontsize':fontsize_subtitle})
fig.text(x = .82, y = subtitle_y, 
         ha = 'center', va = 'bottom',
         s = "Parent w/\nYoung Child".upper(), 
         fontdict={'fontsize':fontsize_subtitle})

# format y-axis 
ax[0].set_yticks(range(0, len(data_tot.index)))
ax[0].set_yticklabels(labels=[wording[k] for k in list(data_yc.index)], 
                      fontsize=fontsize_yticks, color=text_color)
ax[0].tick_params(width = 0)

# remove lines around plot
sns.despine(bottom=True, left=True) 

# dimmensions and customizations for the negative (left) bar plot - TOTAL
ax[0].set_xticks([])
ax[0].set_xlim([-max_xaxis,0]) #set x-axis limits 
ax[0].barh(data_tot.index, data_tot['poor'], color = l_blue)
ax[0].set_title('Poor', ha='right', va='center', fontsize=fontsize_subtitle, x = 0.9)

# create data value labels
for x,y in zip(data_tot['poor'], data_tot.index):
    label = f"{(x*-1)}%"
    ax[0].annotate(label, 
                   (x-nudge_datalab,y), 
                   ha='center', 
                   va='center',
                   fontsize=fontsize_datalabs,
                   weight = 'bold')  

# dimmensions and customizations for the positive (right) bar plot - TOTAL
ax[1].barh(data_tot.index, data_tot['good'], color = d_blue)
ax[1].set_title('Good', ha='left', va='center', fontsize=fontsize_subtitle, x = 0.1)
ax[1].set_xlim([0,max_xaxis]) 
ax[1].set_xticks([]) 
# create data value labels
for x,y in zip(data_tot['good'], data_tot.index):
    label = f"{(x)}%"
    ax[1].annotate(label, 
                 (x+nudge_datalab,y), 
                 ha='center', 
                 va='center',
                 fontsize=fontsize_datalabs,
                 weight = 'bold')

# dimmensions and customizations for the negative (left) bar plot - YOUNG CHILD
ax[2].set_xticks([])
ax[2].set_xlim([-max_xaxis,0]) #set x-axis limits 
ax[2].barh(data_yc.index, data_yc['poor'], color = l_blue)
ax[2].set_title('Poor', ha='right', va='center', fontsize=fontsize_subtitle, x = 0.9)

# create data value labels
for x,y in zip(data_yc['poor'], data_yc.index):
    label = f"{(x*-1)}%"
    ax[2].annotate(label, 
                   (x-nudge_datalab,y), 
                   ha='center', 
                   va='center',
                   fontsize=fontsize_datalabs,
                   weight = 'bold')  

# dimmensions and customizations for the positive (right) bar plot - YOUNG CHILD
ax[3].barh(data_yc.index, data_yc['good'], color = d_blue)
ax[3].set_title('Good', ha='left', va='center', fontsize=fontsize_subtitle, x = 0.1)
ax[3].set_xlim([0,max_xaxis]) 
ax[3].set_xticks([]) 
# create data value labels
for x,y in zip(data_yc['good'], data_yc.index):
    label = f"{(x)}%"
    ax[3].annotate(label, 
                 (x+nudge_datalab,y), 
                 ha='center', 
                 va='center',
                 fontsize=fontsize_datalabs,
                 weight = 'bold')

fig.tight_layout(rect = [0, 0.1, 1, 0.8], pad=0)
    
# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q3. Sample: 2,052 US adults. Dates: August 9-11, 2021. 
See topline questionnaire in appendix for question wording.'''
fig.text(x = 0, y = 0, 
         ha = 'left', va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':fontsize_caption})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.9,0,0.1,0.1], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

# save image file
#plt.show();
fig.savefig('../images/fig05.png', facecolor = "white", dpi=300, bbox_inches='tight');
```
