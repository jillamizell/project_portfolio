```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import matplotlib as mpl
import matplotlib.patches as mpatches
```

```python
rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = 'white'

blue = '#006298'
l_blue = '#99D6EA'
```

```python
data = pd.read_csv('../data/q3_know_exp_df.csv')
data
```

```python
wording = {
    'q3_1_grpd': "Parents who work non-standard\nhours (e.g., third shift)",
    'q3_2_grpd': "Parents who work standard\nhours (e.g., 9 to 5)",
    'q3_3_grpd': "Children",
    'q3_4_grpd': "Childcare workers"
}
```

```python
data_ke = data[data['group'] == 'know_exp'].copy()
data_ke.drop(columns = 'group', inplace = True)
data_ke.set_index('quest', inplace = True)
data_ke = data_ke.reindex(['q3_1_grpd', 'q3_4_grpd', 'q3_3_grpd', 'q3_2_grpd'])
```

```python
data_dke = data[data['group'] == 'dont_know_exp'].copy()
data_dke.drop(columns = 'group', inplace = True)
data_dke.set_index('quest', inplace = True)
data_dke = data_dke.reindex(['q3_1_grpd', 'q3_4_grpd', 'q3_3_grpd', 'q3_2_grpd'])
```

```python
axis_max = 73

fontsize_caption = 16
fontsize_title = 40
fontsize_subtitle = 24
fontsize_ylab = 22
fontsize_datalab = 22
nudge_datalab = 6
range_yaxis = np.array(range(0, len(data_ke.index)))
bar_height = 0.4

# instantiate subplot objects
fig, ax = plt.subplots(ncols = 2, figsize = (13,6), sharey = True)
 
fig.tight_layout(rect = [0, 0.05, 1, 0.95])
#fig.subplots_adjust(wspace = .5)

# plot title
fig.suptitle("Performance of childcare system, by proximity".upper(),
             x = 0.35, y=1,
             fontsize = fontsize_title,
             ha='center', va = 'bottom',
             fontproperties = 'DIN Condensed',
             weight = 'bold')


# remove lines around plot
sns.despine(bottom=True, left=True) 

# format y axis
ax[0].set_yticks(range_yaxis)
ax[0].set_yticklabels(labels=[(wording[k]) for k in list(data_ke.index)], fontsize=fontsize_ylab, color = text_color)

# dimmensions and customizations for the poor (left) bar plot
ax[0].set_xticks([])
ax[0].set_xlim([axis_max * -1,0])
ax[0].set_title("POOR", loc = 'right', fontdict = {'fontsize': fontsize_subtitle, 'fontweight': 'bold'})

ax[0].barh(range_yaxis+bar_height/2, data_ke['poor']*-1, height = bar_height, color = l_blue, label = "Poor")
ax[0].barh(range_yaxis-bar_height/2, data_dke['poor']*-1, height = bar_height, color = l_blue, alpha = 0.7)

# create data value labels
for x,y in zip(data_ke['poor'], range_yaxis):
    label = f"{x}%"
    ax[0].annotate(label, 
                   ((x*-1)+nudge_datalab,y+bar_height/2), 
                   ha='center', va='center',
                   fontsize=fontsize_datalab,
                   color = 'white', weight = 'bold')
for x,y in zip(data_dke['poor'], range_yaxis):
    label = f"{x}%"
    ax[0].annotate(label, 
                   ((x*-1)+nudge_datalab,y-bar_height/2), 
                   ha='center', va='center',
                   fontsize=fontsize_datalab,
                   color = 'white', weight = 'bold')

# dimmensions and customizations for the good (right) bar plot
ax[1].set_xticks([])
ax[1].set_xlim([0, axis_max])
ax[1].set_title("GOOD", loc = 'left', fontdict = {'fontsize': fontsize_subtitle, 'fontweight': 'bold'})

ax[1].barh(range_yaxis+bar_height/2, data_ke['good'], height = bar_height, color = blue, label = "Good")
ax[1].barh(range_yaxis-bar_height/2, data_dke['good'], height = bar_height, color = blue, alpha = 0.7)

# create data value labels
for x,y in zip(data_ke['good'], range_yaxis):
    label = f"{x}%"
    ax[1].annotate(label, 
                 (x-nudge_datalab,y+bar_height/2), 
                 ha='center', va='center',
                 fontsize=fontsize_datalab,
                 color = 'white', weight = 'bold')
for x,y in zip(data_dke['good'], range_yaxis):
    label = f"{x}%"
    ax[1].annotate(label, 
                 (x-nudge_datalab,y-bar_height/2), 
                 ha='center', va='center',
                 fontsize=fontsize_datalab,
                 color = 'white', weight = 'bold')

# add proximity labels
fig.text(x = 0.25, y = 0.825, ha = 'center', s = "High Proximity", fontdict = {'fontsize': fontsize_ylab})
fig.text(x = 0.25, y = 0.75, ha = 'center', s = "Low Proximity", fontdict = {'fontsize': fontsize_ylab})
    
# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q3. Sample: 2,052 US adults. Dates: August 9-11, 2021.
High proximity includes respondents that know someone or have personally experienced any childcare challenges (see Q1).
See topline questionnaire in appendix for question wording.'''
fig.text(x = -0.3, y = 0, 
         ha = 'left', va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':fontsize_caption})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.85,0,0.1,0.1], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

# save image file
plt.show();
#fig.savefig('../images/q3xproximity.png', transparent = True, dpi=300, bbox_inches='tight');
```

```python
title_x = 0.25
#fig_num = 12
y_range = range(0, len(data_ke.index))
y_range_pl = [i+0.2 for i in list(y_range)]
y_range_mi = [i-0.2 for i in list(y_range)]

#################
## CHART SPECS ##
#################
l_blue = "#41B6E6"
d_blue = "#006298"
grey = '#323232'

# font and point sizes, locations
fontsize_ytick = 20
title_size = 30
caption_size = 16
legend_size = 16
fontsize_datalabel = 18
point_size = 2000

# universal axis settings
xmin = 0
xmax = 77
ymin = 4.5
ymax = -0.5  

# create figure and axes
#fig, ax = plt.subplots(ncols = 2, sharey = True, figsize = (10,5))
fig, ax = plt.subplots(figsize = (10,10))

# y-axis settings

ax.set_yticks(y_range) 
ax.set_yticklabels(labels=[(wording[k]) for k in list(data_ke.index)], fontdict={'fontsize': fontsize_ytick})
ax.set_ylim(ymin = ymin, ymax = ymax)
ax.invert_yaxis() 

# Add title
ax.set_title("Performance of childcare system, by proximity".upper(),
             fontsize = title_size,
             fontproperties = 'DIN Condensed',
             x = title_x,
             y = 1)
#                ha = 'center',
#                va = 'center')

################
## KNOW EXPER ##
################

# remove xticks and set x-axis limits
ax.set_xticks([])
ax.set_xlim(xmin = xmin, xmax = xmax)

# The horizontal lines are made using the hline function
ax.hlines(y=y_range_pl, 
             xmin=data_ke.min(axis = 1), 
             xmax=data_ke.max(axis = 1), 
             color='grey', 
             alpha=0.4,
             lw = 5)

# the points showing poor
ax.scatter(data_ke['poor'], 
              y=y_range_pl, 
              color=l_blue, 
              alpha=1, 
              s=point_size,
              label = 'Poor')

# the points showing good
ax.scatter(data_ke['good'], 
              y=y_range_pl, 
              color=d_blue, 
              alpha=1, 
              s=point_size,
              label = 'Good')

# the points showing DK
ax.scatter(data_ke['dk'], 
              y=y_range_pl, 
              color=grey, 
              alpha=1, 
              s=point_size,
              label = "Don't know")



# add data values to each point
for x,y in zip(data_ke['poor'], y_range):
    label = f"{x}%"
    ax.annotate(label, 
                   (x,y+0.2), 
                   ha='center', 
                   va='center',
                   fontsize = fontsize_datalabel,
                   color = 'white',
                   weight = 'bold') 
for x,y in zip(data_ke['good'], y_range):
    label = f"{x}%"
    ax.annotate(label, # this is the text
                   (x,y+0.2),
                   ha='center', # horizontal alignment can be left, right or center
                   va='center',
                   fontsize = fontsize_datalabel,
                   color = 'white',
                   weight = 'bold') 
for x,y in zip(data_ke['dk'], y_range):
    label = f"{x}%"
    ax.annotate(label, # this is the text
                   (x,y+0.2),
                   ha='center', # horizontal alignment can be left, right or center
                   va='center',
                   fontsize = fontsize_datalabel,
                   color = 'white',
                   weight = 'bold') 

######################
## DON'T KNOW EXPER ##
######################
    
# The horizontal lines are made using the hline function
ax.hlines(y=y_range_mi, 
             xmin=data_dke.min(axis = 1), 
             xmax=data_dke.max(axis = 1), 
             color='grey', 
             alpha=0.4,
             lw = 5)

# the points showing poor
ax.scatter(data_dke['poor'], 
              y=y_range_mi, 
              color=l_blue, 
              alpha=0.5, 
              s=point_size,
              label = 'Poor')

# the points showing good
ax.scatter(data_dke['good'], 
              y=y_range_mi, 
              color=d_blue, 
              alpha=0.5, 
              s=point_size,
              label = 'Good')

# the points showing DK
ax.scatter(data_dke['dk'], 
              y=y_range_mi, 
              color=grey, 
              alpha=0.5, 
              s=point_size,
              label = "Don't know")

# add data values to each point
for x,y in zip(data_dke['poor'], y_range):
    label = f"{x}%"
    ax.annotate(label, 
                   (x,y-0.2), 
                   ha='center', 
                   va='center',
                   fontsize = fontsize_datalabel,
                   color = 'white',
                   weight = 'bold') 
for x,y in zip(data_dke['good'], y_range):
    label = f"{x}%"
    ax.annotate(label, # this is the text
                   (x,y-0.2),
                   ha='center', # horizontal alignment can be left, right or center
                   va='center',
                   fontsize = fontsize_datalabel,
                   color = 'white',
                   weight = 'bold') 
for x,y in zip(data_dke['dk'], y_range):
    label = f"{x}%"
    ax.annotate(label, # this is the text
                   (x,y-0.2),
                   ha='center', # horizontal alignment can be left, right or center
                   va='center',
                   fontsize = fontsize_datalabel,
                   color = 'white',
                   weight = 'bold') 

# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q3. Sample: 2,052 US adults. Dates: August 9-11, 2021.
High proximity includes respondents that know someone or have personally experienced any childcare challenges (see Q1).
See topline questionnaire in appendix for question wording.'''
fig.text(x = -0.25, y = 0, 
         ha = 'left',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':caption_size})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([-0.25,-0.1,0.16,0.16], anchor='SW', zorder=1)
newax.imshow(logo)
newax.axis('off')

##fig.subplots_adjust(bottom = 0.4)

# remove borders
sns.despine(bottom=True, left=True)

# legend
ax.legend(ncol = 2, markerscale = 0.25, fontsize = 18, frameon = False, loc = (0.2, 0.8))

fig.text(x = 0.31, y = 0.85, 
         s = "high proximity".upper(), fontdict={'fontsize':caption_size})
fig.text(x = 0.55, y = 0.85, 
         s = "low proximity".upper(), fontdict={'fontsize':caption_size})

# save image file
plt.show();
#fig.savefig(f'../images/fig{fig_num}.png', transparent = True, dpi=300, bbox_inches='tight');
```
