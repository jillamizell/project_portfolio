```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
import matplotlib as mpl
import textwrap
```

```python
rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color
colors = {
    'orange':'#f4731f',
    'turquoise': '#41B6E6',
    'purple': '#512D6D',
    'blue': '#006298',
    'l_blue': '#99D6EA',
    'd_blue': '#01426A',
    'green': '#78BE20'
}
```

```python
data = pd.read_csv('../data/q6_df.csv')
```

```python
data.sort_values('tot', inplace = True)
data.reset_index(drop = True, inplace = True)
data
```

```python
data.iloc[:,3:].max().max()
```

```python
textwrap.fill(data.loc[8]['label'], 50)
```

```python
fig, ax = plt.subplots(ncols = 3, nrows = 3, figsize = (12,15), sharey = True)
fontsize_xlab = 16
fontsize_datalab = 16
fontsize_title = 30
# bar specs
width = 0.25

# format axes
sns.despine(bottom=False, left=True)
fig.tight_layout(h_pad=5, rect = (0, 0.075, 1, 0.95))
ax[0,0].set_yticks([])
ax[0,0].set_ylim([0, 85])
#plt.subplots_adjust(bottom = 0.25)
labels = {'gov':'Government', 'emp':'Employer', 
          'sup_sys':'Individual Support System', 'par_res':"Parents' Responsibility"}
loc_dict = {0:[8,7,6], 1:[5,4,3], 2:[2,1,0]}
y_nudge = 4
for r in range(0,3):
    for c in range(0,3):
        i = loc_dict[r][c]
        ax[r,c].set_xticks([0])
        ax[r,c].set_xticklabels([textwrap.fill(data.loc[i]['label'], 35)], fontsize=fontsize_xlab)
        ax[r,c].bar(0 - width*1.5, data.loc[i]['gov'], width, label=labels['gov'], color = colors['l_blue'])
        ax[r,c].bar(0 - width/2, data.loc[i]['emp'], width, label=labels['emp'], color = colors['turquoise'])
        ax[r,c].bar(0 + width/2, data.loc[i]['sup_sys'], width, label=labels['sup_sys'], color = colors['blue'])
        ax[r,c].bar(0 + width*1.5, data.loc[i]['par_res'], width, label = labels['par_res'], color = colors['purple'])
        ax[r,c].annotate(f"{data.loc[i]['gov']}%", (0 - width*1.5, data.loc[i]['gov']+y_nudge),ha='center', 
                         va='center', fontsize=fontsize_datalab, color = text_color, weight = 'bold')
        ax[r,c].annotate(f"{data.loc[i]['emp']}%", (0 - width/2, data.loc[i]['emp']+y_nudge), ha='center', 
                         va='center', fontsize=fontsize_datalab, color = text_color, weight = 'bold')
        ax[r,c].annotate(f"{data.loc[i]['sup_sys']}%", (0 + width/2, data.loc[i]['sup_sys']+y_nudge), ha='center', 
                         va='center', fontsize=fontsize_datalab, color = text_color, weight = 'bold')
        ax[r,c].annotate(f"{data.loc[i]['par_res']}%", (0 + width*1.5, data.loc[i]['par_res']+y_nudge), ha='center', 
                         va='center', fontsize=fontsize_datalab, color = text_color, weight = 'bold')

# create and format title
title = "Support for employer benefits, by primary support provider"
fig.suptitle(title.upper(),
             fontsize = fontsize_title,
             ha='center', va='bottom',
             fontproperties = 'DIN Condensed')

# create plot legend
handles, labels = ax[0,0].get_legend_handles_labels()
fig.legend(handles, labels, ncol = 4, prop={'size':14}, frameon=False, handlelength = 1, 
           loc = 'upper center', bbox_to_anchor = (0.5,0.98))
# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q5 and Q6. Sample: 2,052 US adults. Dates: August 9-11, 2021. 
See topline questionnaire in appendix for question wording.'''
fig.text(x = 0, y = 0, 
         ha = 'left',
         va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':14})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.9,0,0.1,0.1], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

# save image file
#plt.show();
fig.savefig(f'../images/fig12.png', facecolor = "white", dpi=300, bbox_inches='tight');
```
