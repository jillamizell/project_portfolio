```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
from matplotlib import font_manager
import matplotlib as mpl
import textwrap
```

```python
mpl.rcParams['font.family'] = 'Proxima Nova'
text_color = '#323232'
mpl.rcParams['text.color'] = text_color
mpl.rcParams['axes.labelcolor'] = text_color
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color

colors = {
    'orange':'#f4731f',
    'turqoise': '#41B6E6',
    'purple': '#512D6D',
    'blue': '#006298',
    'l_blue': '#99D6EA',
    'd_blue': '#01426A',
    'green': '#78BE20'
}
```

```python
data = pd.read_csv('../data/q7_df.csv')
data
```

```python
# create plot objects
fig, ax = plt.subplots(figsize = (8,6))

# set some font sizes and label locations as variables
title_size = 40
datalab_size = 28
legtext_size = 22
ytext_size = 28
caption_size = 20
title_x = 0.1

# create bar objects
rects1 = ax.barh([0,1], data[data['Labels'] == 'Disagree']['Prop'], 
                 label = 'Disagree', color = colors['l_blue'], height = 0.6)
rects2 = ax.barh([0,1], data[data['Labels'] == 'Agree']['Prop'], 
                 left = data[data['Labels'] == 'Disagree']['Prop'], 
                 label = 'Agree', color = colors['blue'], height = 0.6)

# create data labels
for x,y in zip(data[data['Labels'] == 'Disagree']['Prop'],[0,1]):
    label = f"{x}%"
    ax.annotate(label, 
                (x/2,y),
                ha='center',
                va='center',
                fontsize=datalab_size,
                color = 'white',
                weight = 'bold') 
for x,y,z in zip(data[data['Labels'] == 'Agree']['Prop'], [0,1], data[data['Labels'] == 'Disagree']['Prop']):
    label = f"{x}%"
    ax.annotate(label, 
                ((x/2)+z,y),
                ha='center',
                va='center',
                fontsize=datalab_size,
                color = 'white',
                weight = 'bold')

# format axes
fig.tight_layout(rect = [0, 0.05, 1, 0.9])
sns.despine(bottom=True, left=True)
ax.set_xticks([])

# format y-axis
ax.invert_yaxis()
ax.set_yticks([0,1])
ax.set_yticklabels([textwrap.fill('Improving the childcare and education system so that parents can stay in the workforce',50),
                    textwrap.fill('Improving job quality and pay so that families can support themselves with one income and a parent can stay home and provide care',50)], 
                    fontdict={'fontsize':ytext_size})


# create title
fig.suptitle('to improve childcare in the u.s. efforts should focus on:'.upper(),
             x = 0,
             fontsize = title_size,
             ha='center', va = 'bottom',
             fontproperties = 'DIN Condensed')

# create legend
fig.legend(ncol=2, loc='upper center', frameon=False, prop={'size':legtext_size}, bbox_to_anchor = (0.5,1))

# create caption
captext = '''Source: JUST Capital/Harris Omnibus Poll Q7. Sample: 2,052 US adults. Dates: August 9-11, 2021.
See topline questionnaire in appendix for question wording.'''
fig.text(x = -1, y = 0, 
         ha = 'left', va = 'bottom',
         s = captext, fontdict={'family':'Proxima Nova Condensed', 'fontsize':caption_size})

# insert logo image
logo = plt.imread('../assets/Just Capital_Logo_RGB_500px_SSf.png')
newax = fig.add_axes([0.88,0,0.15,0.15], anchor='SE', zorder=1)
newax.imshow(logo)
newax.axis('off')

# save image file
plt.show()
#fig.savefig(f'../images/fig{fig_num}.png', transparent = True, dpi=300, bbox_inches='tight');
```
